```python

text = """
Nature is an incredible force that shapes the world around us.
The mountains stand tall and proud, reaching towards the sky.
The rivers flow gently through the valleys, nourishing the land.
The forests are full of life, with trees that touch the clouds.
The oceans are vast and deep, hiding mysteries beneath the waves.
Animals roam freely across the plains and forests.
The sun rises every morning, bringing warmth and light to the world.
The moon glows softly at night, guiding travelers on their journey.
Stars twinkle in the dark sky, reminding us of the vast universe.
Nature teaches us patience, beauty, and the importance of balance.
The wind whispers through the leaves, carrying stories from distant lands.
Rain falls gently on the earth, giving life to plants and trees.
Every season brings its own beauty and wonder to the natural world.
"""

print(" Text loaded successfully!")
print(f"Total characters: {len(text)}")
print(f"Total words: {len(text.split())}")
```

```python
import markovify


model = markovify.Text(text, state_size=2)

print(" Markov Chain model built successfully!")
```

```python
print(" Generated Sentences:\n")


for i in range(10):
    sentence = model.make_sentence()
    if sentence:
        print(f"Sentence {i+1}: {sentence}")
```

```python

text2 = """
The city lights shine brightly across the dark horizon every single night.
People walk through busy streets filled with laughter and joy always.
Technology has changed the way we live, work, and communicate forever.
Computers process millions of calculations every single second without stopping.
Artificial intelligence is transforming industries across the entire modern world.
Students learn new skills every day to prepare for the future ahead.
Books open doors to new worlds, ideas, and endless imagination for readers.
Music fills the air with emotions that words alone cannot express easily.
Artists paint beautiful pictures that capture moments frozen in time forever.
Science explores the unknown, pushing boundaries of human knowledge every day.
The future belongs to those who believe in the beauty of their dreams.
Hard work and dedication always lead to success in the long run.
Kindness and compassion make the world a better place for everyone always.
Every new day brings new opportunities, new challenges, and new beginnings ahead.
Dreams inspire people to achieve greatness beyond what they thought was possible.
"""


combined_text = text + text2


model2 = markovify.Text(combined_text, state_size=2)

print(" Bigger model built!")
print(f"Total words now: {len(combined_text.split())}")
```

```python
print(" Generated Sentences from Bigger Model:\n")


for i in range(10):
    sentence = model2.make_sentence()
    if sentence:
        print(f"Sentence {i+1}: {sentence}")

print("\n-----------------------------------\n")


print(" Sentences starting with 'The':\n")
for i in range(5):
    sentence = model2.make_sentence_with_start("The")
    if sentence:
        print(f"Sentence {i+1}: {sentence}")
```

```python
print("=" * 50)
print("   TASK 03 - TEXT GENERATION WITH MARKOV CHAINS")
print("=" * 50)

print("\n Model Details:")
print(f"   Training text words: {len(combined_text.split())}")
print(f"   State size: 2 (looks at 2 previous words)")

print("\n Sample Generated Sentences:")
for i in range(5):
    sentence = model2.make_sentence()
    if sentence:
        print(f"   {i+1}. {sentence}")

print("\n Custom Start Word Sentences ('The'):")
for i in range(3):
    sentence = model2.make_sentence_with_start("The")
    if sentence:
        print(f"   {i+1}. {sentence}")

print("\n Task 03 Completed Successfully!")
print("=" * 50)
```

```python

with open("markov_output.txt", "w") as f:
    f.write("TASK 03 - TEXT GENERATION WITH MARKOV CHAINS\n")
    f.write("=" * 50 + "\n\n")

    f.write("Generated Sentences:\n\n")
    for i in range(10):
        sentence = model2.make_sentence()
        if sentence:
            f.write(f"{i+1}. {sentence}\n")

print(" Output saved to markov_output.txt")
```

```python

test_words = ["The", "Nature", "Animals", "Stars", "Kindness", "Science", "Dreams"]

print(" TESTING MARKOV CHAIN\n")
print("=" * 50)

for word in test_words:
    try:
        sentence = model2.make_sentence_with_start(word)
        if sentence:
            print(f" Start: '{word}'")
            print(f"   → {sentence}\n")
    except:
        print(f" Start: '{word}'")
        print(f"   →  Couldn't generate (word not in training text)\n")
```
